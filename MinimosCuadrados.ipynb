```python
import numpy as np
from ipywidgets import interact
import matplotlib.pyplot as plt

p1 = (5.4, 3.2)
p2 = (9.5, 0.7)
p3 = (12.3, -3.6)

def calculate_least_squares_line(points):
    x_coords, y_coords = zip(*points)
    m, b = np.polyfit(x_coords, y_coords, 1)
    return m, b

def update_plot(p2_x, p2_y):
    points = [p1, (p2_x, p2_y), p3]
    m, b = calculate_least_squares_line(points)
    x_coords = [point[0] for point in points]
    y_coords = [point[1] for point in points]
    x_line = np.linspace(min(x_coords), max(x_coords), 100)
    y_line = m * x_line + b
    plt.figure(figsize=(10, 6))
    plt.scatter(x_coords, y_coords, color="red", label="Puntos")
    plt.plot(x_line, y_line, color="blue", label=f"y = {m:.2f}x + {b:.2f}")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Lnea de Mnimos Cuadrados y Puntos")
    plt.legend()
    plt.grid()
    plt.show()
    print(f"Pendiente (m): {m:.2f}, Ordenada al origen (b): {b:.2f}")

_ = interact(update_plot, p2_x=(5.5, 12.3, 0.1), p2_y=(-10.0, 10.0, 0.1))

def get_least_squares_params(points):
    return calculate_least_squares_line(points)

example_points = [p1, p2, p3]
m, b = get_least_squares_params(example_points)
print(f"Pendiente (m): {m:.2f}, Ordenada al origen (b): {b:.2f}")
```
